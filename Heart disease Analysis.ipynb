```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
file_path = "heart_attack_youngsters_india.csv"
data = pd.read_csv(file_path)
```

```python
data_info = data.info()
data_head = data.head()

data_info, data_head
```

```python
summary_statistics = data.describe()


missing_values = data.isnull().sum()

summary_statistics, missing_values
```

```python

# Prepare data for predictive modeling
X = data[['Age', 'Cholesterol Levels (mg/dL)', 'BMI (kg/m²)', 'Screen Time (hrs/day)', 'Sleep Duration (hrs/day)']]  # Example features
y = data['Heart Attack Likelihood'].apply(lambda x: 1 if x == 'Yes' else 0)  # Encode target variable

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluation
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
